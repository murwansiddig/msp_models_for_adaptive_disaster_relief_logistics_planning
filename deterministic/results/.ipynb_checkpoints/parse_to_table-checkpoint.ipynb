```julia
using CSV, DataFrames
```

```julia
raw = CSV.read("saveme.csv",DataFrame);
data = Matrix(raw);
```

```julia
rho = ["0.001", "0.60", "5.00"];
SPs = ["3", "6", "9"]
DPs = ["10", "20", "30"];
```

```julia
'%' in data[1,12]
```

```julia
data[1,12][1:end-1]*"\\%"
```

```julia
count = 0;


for r in rho
    for s in SPs
        for d in DPs
            count += 1
            #first 3 terms
            term1 = ""
            if count in [1, 10, 19]
                term1 *= "$r & $s \$\\quad\$       $d && "
            elseif count in [4, 7, 13, 16, 22, 25]
                term1 *= "   & $s \$\\quad\$       $d && "
            else
                term1 *= "   &    \$\\quad\\quad\$ $d && "
            end
            print(term1)
            ccount = 0
            for i=1:12
                ccount += 1
                if '%' âˆ‰ data[count,i]
                    if ccount < 12
                        print(data[count,i], " &")
                    else
                        print(data[count,i], " ")
                    end
                else
                    if ccount < 12
                        print(data[count,i][1:end-1]*"\\%", " &")
                    else
                        print(data[count,i][1:end-1]*"\\%", " ")
                    end                    
                end
                if ccount in [4, 8]
                    print("& ")
                else
                    print(" ")
                end
            end
            print(" \\\\")
            println(" ")
        end
    end
    print(" \\\\")
    println(" ")
end
```
